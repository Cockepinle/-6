Цель работы: на данной практической работе необходимо использовать датасет с прошлой практической работы и визуализировать данные в виде графиков. Необходимо использовать: Линейный график - минимум 3 штуки (можно использовать наложение графиков), cтолбчатая диаграмма- минимум 3 штуки (можно также использовать наложение), круговая диаграмма - минимум 2 штуки, диаграмма рассеяния - минимум 2 штуки, гистограмма - минимум 3 штуки

```python
import pandas as pd

waterQuality1 = pd.read_csv("waterQuality1.csv")
waterQuality1
```

```python
import matplotlib.pyplot as plt
x = waterQuality1["aluminium"].head()
y = waterQuality1["ammonia"].head()
plt.plot (x, y, label = 'Линейный график')
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid()
plt.legend()
plt.show()
```

1. Линейный график, который содержит в себе соотношение в воде значений колонок алюминий и аммиак, по оси х это алюминий, а по оси у аммиак, т.е. к примеру, при значении 1.0 алюминия, аммиак имеет значение 14.02.

```python
x = waterQuality1["arsenic"].head()
y = waterQuality1["barium"].head()
plt.plot (x, y, color = 'green', marker = 'o', label = 'Линейный график')
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid()
plt.legend()
plt.show()
```

1. Линейный график, который содержит в себе соотношение в воде наличение значений колонок мышьяк и барий, по оси х это мышьяк, а по оси х это барий, т.е. к примеру, при содержании 0.010 мышьяка, бариум составляет  приблизительно 3.7.

```python
import matplotlib.pyplot as plt
x = waterQuality1["cadmium"].head()
y = waterQuality1["chloramine"].head()
plt.plot (x, y, color = 'pink', marker = 'o', markersize = 7, label = 'Линейный график')
plt.xlabel('Ось х')
plt.ylabel('Ось y')
plt.grid()
plt.legend()
plt.show()
```

1. Линейный график, который содержит в себе соотношение в воде наличение значений колонок кадмий и хлорамин, где ось х отвечает за значения камдия, а ось у отвечает за значения хлорамина в воде, к примеру, при значении 0.001 камдия, хлорамин составляет на графике 7.5.

```python
x = waterQuality1["chromium"].head()
y = waterQuality1["copper"].head()
plt.bar (x, y, label = 'Соотношение количества в воде хрома и медь', color = "pink")
plt.xlabel('Хром')
plt.ylabel('Медь')
plt.title('Пример столбчатой диаграммы')
plt.legend()
plt.show()
```

2. В данном графике отображает соотношение в воде количества меди (ось у) и хрома (ось х) для их сравнения, к примеру, от -0.4 до 0.4 количества хрома, тогда количество меди составяет 1.6.

```python
x = waterQuality1["flouride"].head()
y = waterQuality1["bacteria"].head()
plt.bar (x, y, label = 'Соотношение количества в воде фторида и бактерий', color = "black")
plt.xlabel('Фторид')
plt.ylabel('Бактерии')
plt.title('Пример столбчатой диаграммы фторид/бактерии')
plt.legend()
plt.show()
```

2. В данном графике отображает соотношение в воде количества бактерий (ось у) и фторида (ось х) для их сравнения, то есть там где количество фторида меньше, там и количествотбактерий тоже меньше, к примеру, от -0.25 до 0.48 фторида, тогда бактерии составляют 0.2.

```python
x = waterQuality1["lead"].head()
y = waterQuality1["nitrates"].head()
plt.bar (x, y, label = 'Соотношение количества в воде свинца и нитратов', color = "purple")
plt.xlabel('Свинец')
plt.ylabel('Нитраты')
plt.title('Пример столбчатой диаграммы фторид/бактерии')
plt.legend()
plt.show()
```

2. В данном графике отображает соотношение в воде количества нитраты (ось у) и свинец (ось х) для их сравнения, к примеру, при значении -0.4 свинца, значение нитратов составляет 1.8.

```python
perchlorate_counts = waterQuality1['perchlorate'].tail(10).value_counts()
plt.pie(perchlorate_counts, labels = perchlorate_counts.index, autopct ='%1.1f%%', 
colors = ['pink', 'orange'])
plt.title('Соотношение перхлората и его значений в воде')
plt.show()
```

3. В данной круговой диаграмме представлены 10 возможных знаяений перхлората в воде и их процент, который используется в датасете, к примеру, значение 1.35 находится в воде 20%.

```python
radium_counts = waterQuality1['radium'].tail(10).value_counts()
plt.pie(radium_counts, labels = radium_counts.index, autopct ='%1.1f%%', 
colors = ['pink', 'black'])
plt.title('Соотношение радия и его значений в воде')
plt.show()
```

3. В данной круговой диаграмме представлены 10 возможных знаяений радий в воде и их процент, который используется в датасете, к примеру, значение 1.73 находится в воде 10%.

```python
y = waterQuality1['selenium'][10:16]
x = waterQuality1['uranium'][10:16]
plt.scatter(x,y)
plt.show()
```

4. На данной рассеянной диаграме можно увидеть кол-во соотношение между селен (ось у) и ураном (ось х), к примеру, при значении 0.03 урана, селен составляет 0.02.

```python
y = waterQuality1['radium'][10:15]
x = waterQuality1['arsenic'][10:15]
plt.scatter(x,y)
plt.show()
```

4. На данной рассеянной диаграме можно увидеть кол-во соотношение между радий (ось у) и мышьяк (ось х), к примеру, при значении 0.02 радия, мышьяк составляет 2.1.

```python
plt.figure(figsize=(7,7))
plt.hist(waterQuality1['copper'], bins = 30, color =["black"],
 edgecolor='pink',alpha=0.7, label ='Колчество меди в составе воды')
plt.title("Гистограмма")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(12,12))
plt.hist(waterQuality1['chromium'], bins = 30, color =["green"],
 edgecolor='purple',alpha=0.7, label ='Колчество хрома в составе воды')
plt.title("Гистограмма")
plt.legend()
plt.show()
```
